# Import respective Libraries

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("lung_cancer.csv")
df
```

# Check the null Values

```python
df.info()
```

# Drop the unwanted columns

```python
# List of columns to drop
columns_to_drop = ['PATIENT_ID','ADOPTED_STATUS','NUMBER_OF_PREGNANCIES','MONTH_OF_BIRTH','YEAR_OF_BIRTH','Constant_Exhaustion']

# Filter out columns that are not in the DataFrame
existing_columns_to_drop = [col for col in columns_to_drop if col in df.columns]

# Drop the existing columns
df.drop(existing_columns_to_drop, axis=1, inplace=True)
```

```python
df['GENOMIC SEX'] = df['GENOMIC SEX'].str.replace('FEMALE','F',regex=True)
df['GENOMIC SEX'] = df['GENOMIC SEX'].str.replace('MALE','M',regex=True)
```

```python
df
```

```python
# Define mappings
sex_mapping = {'M': 1, 'F': 0}
blood_type_mapping = {'A': 0, 'B': 1, 'AB': 2, 'O': 3}
yes_no_mapping = {'YES': 2, 'NO': 1}

# Apply mappings
df['GENOMIC SEX'] = df['GENOMIC SEX'].map(sex_mapping)
df['BLOOD_TYPE'] = df['BLOOD_TYPE'].map(blood_type_mapping)
#df['Constant_Exhaustion'] = df['Constant_Exhaustion'].map(yes_no_mapping)
df['LUNG_CANCER'] = df['LUNG_CANCER'].map(yes_no_mapping)
```

```python
df
```

# Categorical to numerical

```python
# Convert columns to numeric
df['DAILY_CIGARETTES'] = pd.to_numeric(df['DAILY_CIGARETTES'], errors='coerce')
df['COPD_DIAGNOSES'] = pd.to_numeric(df['COPD_DIAGNOSES'], errors='coerce')
df['WHEEZING'] = pd.to_numeric(df['WHEEZING'], errors='coerce')
df['SHORTNESS_OF_BREATH'] = pd.to_numeric(df['SHORTNESS_OF_BREATH'], errors='coerce')
```

# Missing values filling

```python
# Fill missing values
df['DAILY_CIGARETTES'].fillna(df['DAILY_CIGARETTES'].median(), inplace=True)
df['COPD_DIAGNOSES'].fillna(df['COPD_DIAGNOSES'].median(), inplace=True)
df['WHEEZING'].fillna(df['WHEEZING'].median(), inplace=True)
df['SHORTNESS_OF_BREATH'].fillna(df['SHORTNESS_OF_BREATH'].median(), inplace=True)
```

```python
df
```

```python
df.info()
```

```python
df['NUMBER_OF_SIBLINGS'].fillna(0,inplace=True)
```

```python
# Step 1: Convert non-numeric values to NaN
df['AGE'] = pd.to_numeric(df['AGE'], errors='coerce')

# Step 2: Calculate the mean of the 'AGE' column, excluding NaN values
mean_age = df['AGE'].mean()

# Step 3: Fill NaN values with the mean_age
df['AGE'].fillna(mean_age, inplace=True)

# Step 4: Convert the 'AGE' column to integer type
df['AGE'] = df['AGE'].astype(int)
```

```python
df.info()
```

```python
df
```

```python
# Handle missing values in 'BLOOD_TYPE'
# Option 1: Fill missing values with a specific value or use a placeholder
df['BLOOD_TYPE'].fillna(0, inplace=True)

# Option 2: Convert 'BLOOD_TYPE' to categorical
df['BLOOD_TYPE'] = df['BLOOD_TYPE'].astype('object')

# Handle missing values in 'LUNG_CANCER'
# Option 1: Fill with a placeholder or mode
df['LUNG_CANCER'].fillna(0, inplace=True)

# Convert 'LUNG_CANCER' to categorical
df['LUNG_CANCER'] = df['LUNG_CANCER'].astype('object')
```

```python
df.info()
```

```python
# Find the mode of the 'GENOMIC SEX' column
mode_genomic_sex = df['GENOMIC SEX'].mode()[0]

# Fill missing values with the mode
df['GENOMIC SEX'].fillna(mode_genomic_sex, inplace=True)

# Option 2: Convert 'BLOOD_TYPE' to categorical
df['GENOMIC SEX'] = df['GENOMIC SEX'].astype('object')
```

```python
# Print current columns to verify
print("Columns in DataFrame:", df.columns)

# Trim any leading or trailing spaces from column names
df.columns = df.columns.str.strip()

# Verify the updated column names
print("Updated Columns in DataFrame:", df.columns)

# Fill missing values in 'FATIGUE' column
df['FATIGUE'].fillna(0, inplace=True)

# Print DataFrame to see changes
print("Updated DataFrame:")
print(df)
```
